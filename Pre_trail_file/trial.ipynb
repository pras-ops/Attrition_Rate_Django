```python
# data_transformation.py

from sklearn.base import BaseEstimator, TransformerMixin

class GenderTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.gender_dict = None

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        if self.gender_dict is None:
            self.gender_dict = {'male', 'female'}  

        def Gen(x):
            if x in self.gender_dict:
                return str(x)
            else:
                return 'other'
        
        data_l = X["Gender "].apply(Gen)
        X['New Gender'] = data_l
        return X
class PromotedTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        def Promoted(x):
            if x == 'Promoted':
                return int(1)
            else:
                return int(0)

        data_l = X["Promoted/Non Promoted"].apply(Promoted)
        X['New Promotion'] = data_l
        return X

import pandas as pd
#from data_transformation import GenderTransformer  # Uncomment this line
from sklearn.pipeline import Pipeline

# Step 1: Read the CSV file into the DataFrame
attrdata = pd.read_csv("Table_1.csv")

# Step 2: Create the pipeline
data_pipeline = Pipeline([
    ('gender_transform', GenderTransformer()),
    ('promotion_transform', PromotedTransformer())
])

# Step 3: Apply the pipeline to the DataFrame
transformed_attrdata = data_pipeline.transform(attrdata)

# Now, transformed_attrdata contains the DataFrame after applying the gender transformation and dropping the specified columns (if there were any in the original transformer, which is not present in this version).
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
import pandas as pd

class DataTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = self._promoted_transform(X)
        X = self._data_drop(X)
        return X

    def _promoted_transform(self, X):
        def Promoted(x):
            if x == 'Promoted':
                return int(1)
            else:
                return int(0)

        data_l = X["Promoted/Non Promoted"].apply(Promoted)
        X['New Promotion'] = data_l
        return X

    def _data_drop(self, X):
        # Drop the specified columns from the DataFrame
        X.drop(["table id", "name", "Marital Status", "Promoted/Non Promoted", "Function", "Emp. Group",
                "Job Role Match", "Location", "Hiring Source", "Gender ", 'Tenure', 'New Gender', 'New Marital', 'New EMP'],
               axis=1, inplace=True)
        return X
    
import pandas as pd
#from data_transformation import GenderTransformer  # Uncomment this line
from sklearn.pipeline import Pipeline

# Step 1: Read the CSV file into the DataFrame
attrdata = pd.read_csv("Table_1.csv")

# Step 2: Create the pipeline
data_pipeline = Pipeline([
    ('gender_transform', GenderTransformer()),
    ('promotion_transform', PromotedTransformer())
])

# Step 3: Apply the pipeline to the DataFrame
transformed_attrdata = data_pipeline.transform(attrdata)

```
